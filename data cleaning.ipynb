Clean empty Cells


```python
import pandas as pd
df = pd.read_excel('sample_data/Customer Call List.xlsx')
print(df)
new_df = df.dropna()
new_df
```

Remove Duplicate Rows

```python
#Show List
print(df.duplicated())

#Assign data frame to a new data frame
new_df = df

#Drop Duplicate Records
new_df.drop_duplicates(inplace = True)

#Display with Removed Records
new_df
```

Drop useless column

```python
new_df = new_df.drop(columns = "Not_Useful_Column")
new_df
```

Standardise Yes to Y and No to N

```python
new_df["Paying Customer"] = new_df["Paying Customer"].str.replace('Yes','Y')
new_df["Paying Customer"] = new_df["Paying Customer"].str.replace('No','N')
```

```python
new_df["Do_Not_Contact"] = new_df["Do_Not_Contact"].str.replace('Yes','Y')
new_df["Do_Not_Contact"] = new_df["Do_Not_Contact"].str.replace('No','N')
```

```python
new_df
```

Clean irregular Characters in Last Name

```python
new_df["Last_Name"] = new_df["Last_Name"].str.lstrip("...")
new_df["Last_Name"] = new_df["Last_Name"].str.lstrip("/")
new_df["Last_Name"] = new_df["Last_Name"].str.rstrip("_")
new_df["Last_Name"] = new_df["Last_Name"].str.strip("123._/")
new_df
```

Clean phone numbers from irregular characters

```python
new_df["Phone_Number"] = new_df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
new_df["Phone_Number"] = new_df["Phone_Number"].apply(lambda x: str(x))
new_df["Phone_Number"] = new_df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
new_df["Phone_Number"] = new_df["Phone_Number"].str.replace('nan--','')
new_df["Phone_Number"] = new_df["Phone_Number"].str.replace('Na--','')
new_df["Phone_Number"] = new_df["Phone_Number"].str.replace('-','')
new_df["Phone_Number"] = new_df["Phone_Number"].str.replace('|','')
new_df["Phone_Number"] = new_df["Phone_Number"].str.replace('/','')
new_df
```

Split Address into 3 Columns

```python
new_df[['Street_Address', 'State', 'Zip_Code']] = new_df['Address'].str.split(',', expand=True)
new_df = new_df.drop(columns = 'Address')
new_df
```

Remove all rows that do not have a phone number

```python
for x in new_df.index:
    if new_df.loc[x, "Do_Not_Contact"] == 'Y':
        new_df.drop(x, inplace=True)
new_df
```

```python
for x in new_df.index:
    if new_df.loc[x, "Phone_Number"] == '' or new_df.loc[x, "Phone_Number"] == 'Na' :
        new_df.drop(x, inplace=True)
new_df
```
